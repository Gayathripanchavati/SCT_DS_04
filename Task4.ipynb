```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of accidents by weather
plt.figure(figsize=(10, 6))
sns.countplot(data=df_accidents, x='Weather', palette='viridis')
plt.title('Distribution of Accidents by Weather Condition')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Accidents')
plt.show()

# Distribution of accidents by hour
plt.figure(figsize=(12, 6))
sns.countplot(data=df_accidents, x='Hour', palette='viridis')
plt.title('Distribution of Accidents by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Accidents')
plt.show()

# Relationship between Fatalities and Injuries
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_accidents, x='Fatalities', y='Injuries', alpha=0.6)
plt.title('Relationship between Fatalities and Injuries')
plt.xlabel('Fatalities')
plt.ylabel('Injuries')
plt.show()

# Relationship between Fatalities and Cause
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_accidents, x='Cause', y='Fatalities', palette='viridis')
plt.title('Fatalities by Accident Cause')
plt.xlabel('Cause')
plt.ylabel('Fatalities')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Relationship between Injuries and Cause
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_accidents, x='Cause', y='Injuries', palette='viridis')
plt.title('Injuries by Accident Cause')
plt.xlabel('Cause')
plt.ylabel('Injuries')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import dash
from dash import dcc
from dash import html
import plotly.express as px

# Recreate visualizations using Plotly Express
fig_weather = px.histogram(df_accidents, x='Weather', title='Distribution of Accidents by Weather Condition')
fig_hour = px.histogram(df_accidents, x='Hour', title='Distribution of Accidents by Hour of the Day')
fig_fatalities_injuries = px.scatter(df_accidents, x='Fatalities', y='Injuries', title='Relationship between Fatalities and Injuries')
fig_fatalities_cause = px.box(df_accidents, x='Cause', y='Fatalities', title='Fatalities by Accident Cause')
fig_injuries_cause = px.box(df_accidents, x='Cause', y='Injuries', title='Injuries by Accident Cause')


app = dash.Dash(__name__)

app.layout = html.Div(children=[
    html.H1(children='Accident Data Dashboard'),

    html.Div(children='''
        Dashboard visualizing accident data.
    '''),

    html.Div([
        html.Div([
            dcc.Graph(
                id='weather-distribution',
                figure=fig_weather
            ),
        ], style={'width': '49%', 'display': 'inline-block'}),

        html.Div([
            dcc.Graph(
                id='hour-distribution',
                figure=fig_hour
            ),
        ], style={'width': '49%', 'display': 'inline-block', 'float': 'right'}),
    ]),

    html.Div([
        html.Div([
            dcc.Graph(
                id='fatalities-injuries-relationship',
                figure=fig_fatalities_injuries
            ),
        ], style={'width': '49%', 'display': 'inline-block'}),

        html.Div([
            dcc.Graph(
                id='fatalities-cause-distribution',
                figure=fig_fatalities_cause
            ),
        ], style={'width': '49%', 'display': 'inline-block', 'float': 'right'}),
    ]),

    html.Div([
        html.Div([
            dcc.Graph(
                id='injuries-cause-distribution',
                figure=fig_injuries_cause
            ),
        ], style={'width': '49%', 'display': 'inline-block'}),
    ]),
])

if __name__ == '__main__':
    app.run(debug=True, mode='inline')
```
